# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time
import random
from collections import defaultdict


class kumanda:
    def __init__(self):
        self.kanallar = {1: "TRT ÇOCUK", 2: "MİNİKA GO",
                         3: "MİNİKA GO", 4: "CARTOON NETWORK",
                         5: "DİSNEY CHANNEL", 6: "NİCKELODEON",
                         7: "PLANET ÇOCUK", 8: "BABYTV", 9: "BBC NEWS",
                         11: "NATİONAL GEOGRAPHİC", 12: "NOW TV"}
        self.tuslar = {
            "p": "TV Aç",
            "q": "TV KAPAT",
            "+": "Ses Aç",
            "-": "Ses Kıs",
            "m": "Sessize Al/Sessizden Çık", # Açıklama güncellendi
            "u": "Kanal Yukarı",
            "d": "Kanal Aşağı",
            "ok": "Onayla / Seç",
            "esc": "Çık / Geri Dön",
            "1-9": "Kanal Numarası",
            "s": "Kaynak Seç (HDMI/AV/USB)",
            "i": "Bilgi Göster",  # Bilgi tuşu
            "g": "Yayın Rehberi",
            "v": "Kanal sayısı öğrenme",
            "r": "Rastegele kanala geç",
            "y": "Kanal Ekle",
            "b": "Mavi Tuş",
            ">": "Oynat (▶)",
            "||": "Durdur (⏸)",
            ">>": "İleri Sar (⏩)",
            "<<": "Geri Sar (⏪)"
        }

        self.volume = 30  # Başlangıç ses seviyesi
        self.tv_acik = False # DÜZELTME: self kullanıldı
        self._onceki_volume = 30 # Sessize almadan önceki ses seviyesini saklamak için (isteğe bağlı)
        self.mevcut_kanal_numarasi = 1 # Başlangıç kanalı (isteğe bağlı)


    def menu(self):
        # Düzeltme: Menü sadece TV açıkken gösterilir
        if self.tv_acik:
            print("""---------------MENU--------------\n

                YAPILABİLEN İŞLEMLER""")
            for tus, aciklama in self.tuslar.items():
                print(f"            '{tus}': {aciklama}")
            print("---------------------------------")
        else:
             print("Televizyon kapalı. Menüyü görmek için TV'yi açın.")


    def run(self):
        print("Televizyon kumandası başlatıldı. Çıkmak için 'q' tuşuna basın.")

        while True:
            self.islem_girisi = input("Yapmak istediğiniz işlemi tuşları kullanarak gerçekleştirin: ").lower()

            # Kullanıcı çıkmak istedi mi? (TV açık veya kapalı olabilir)
            if self.islem_girisi == 'q':
                print("Televizyon kapatılıyor.")
                self.tv_acik = False # DÜZELTME: self kullanıldı
                break # Ana while döngüsünü kırar

            action_performed = False  # Döngünün her adımının başında sıfırlanır

            # Televizyon KAPALIYKEN yapılacak işlemler
            if not self.tv_acik: # DÜZELTME: self kullanıldı
                if self.islem_girisi == "p":
                    print("Televizyon açılıyor.")
                    self.tv_acik = True # DÜZELTME: self kullanıldı
                    self.menu() # TV açılınca menüyü göster
                    action_performed = True # 'p' tuşu geçerli bir işlem oldu
                else:
                    # TV kapalıyken 'q' ve 'p' dışında bir tuşa basıldıysa
                    print("Televizyon kapalı. Lütfen önce 'p' tuşuna basarak televizyonu açın veya çıkmak için 'q' tuşuna basın.")

                # TV kapalıyken 'p' dışında bir tuşa basıldıysa, döngünün geri kalanını atla
                if not action_performed: # Eğer 'p' tuşu ile TV açılmadıysa
                    continue # Bu adımda başka işlem yapmadan döngünün başına dön

            # Buraya gelindiyse, TV ya yeni açılmıştır ('p' tuşu ile) ya da zaten AÇIKTIR.
            # TV AÇIKKEN yapılacak işlemler (action_performed True ise yeni açılmıştır, False ise zaten açıktır)

            # Ses işlemleri
            if self.islem_girisi in ('+', '-','m'):
                self.ses()
                action_performed = True # Ses işlemi yapıldı

            # Kanal ekleme
            elif self.islem_girisi == 'y':
                self.kanal_ekleme()
                action_performed = True

            # Kanal sayısı öğrenme
            elif self.islem_girisi == 'v':
                self.kanal_sayisi()
                action_performed = True

            # Rastgele kanala geçme
            elif self.islem_girisi == 'r':
                self.random_kanal()
                action_performed = True

            # Televizyon Bilgileri
            elif self.islem_girisi == 'i':
                self.bilgi()
                action_performed = True

            # Kanal değiştirme (yukarı/aşağı) - Bu metodları eklemeniz gerekir
            # elif self.islem_girisi == 'u':
            #      self.kanal_yukari()
            #      action_performed = True
            # elif self.islem_girisi == 'd':
            #      self.kanal_asagi()
            #      action_performed = True

            # Sayısal kanal girişi - Bu metodu eklemeniz gerekir
            elif self.islem_girisi.isdigit():
            
                self.kanal_degistir(self.mevcut_kanal_numarasi)
                action_performed = True

            # ... Diğer TV açıkken yapılacak işlemler için elif blokları buraya eklenecek
            # Örneğin: 'ok', 'esc', 's', 'g', 'b', '>', '||', '>>', '<<' tuşları

            # Eğer TV açıksa ve yukarıdaki geçerli işlemlerden hiçbiri yapılmadıysa
            if self.tv_acik and not action_performed: # DÜZELTME: self kullanıldı
                print("Geçersiz işlem! Lütfen belirtilen seçeneklerden birini girin.")
                # Kullanıcıya sadece TV açıkken kullanabileceği tuşları göstermek daha iyi olabilir
                # print("Mevcut tuşlar:", self.tuslar) # Tüm tuşları göstermek yerine


    def ses(self):
        # DÜZELTME: Ses kısma ve sessizden çıkma eklendi, 0-100 kontrolü yapıldı
        if self.islem_girisi == "m":
            # Eğer ses zaten 0 değilse sessize al
            if self.volume != 0:
                print(f"Ses {self.volume} iken sessize alındı.")
                self._onceki_volume = self.volume # Sessize almadan önceki sesi sakla
                self.volume = 0
                print("Televizyon sessize alınmıştır.")
            else:
                # Eğer ses zaten 0 ise ve 'm'ye basıldıysa, sesi geri aç (önceki seviyeye veya varsayılana)
                if hasattr(self, '_onceki_volume') and self._onceki_volume > 0: # _onceki_volume var mı ve 0'dan büyük mü kontrol et
                     self.volume = self._onceki_volume
                     print(f"Sessizden çıkıldı. Ses: {self.volume}")
                     # del self._onceki_volume # İsteğe bağlı: Saklanan değeri silebilirsiniz
                else:
                     # Eğer ilk kez sessize alınıyorsa veya _onceki_volume 0 ise/silinmişse, varsayılan bir değere dön
                     self.volume = 20 # Örneğin 20'ye ayarla
                     print(f"Sessizden çıkıldı. Ses: {self.volume}")


        elif self.islem_girisi == "+":
            if self.volume < 100:
                self.volume += 10
                if self.volume > 100: self.volume = 100 # 100'ü geçmemesini sağla
                print(f"Ses: {self.volume}")
            else:
                print("Şuan zaten en yüksek ses seviyesindesiniz (100).")

        elif self.islem_girisi == "-": # SES KISMA EKLENDİ
            if self.volume > 0:
                self.volume -= 10
                if self.volume < 0: self.volume = 0 # 0'ın altına düşmemesini sağla
                print(f"Ses: {self.volume}")
            else:
                print("Şuan zaten en düşük ses seviyesindesiniz (0).")


    def kanal_ekleme(self):
        # Düzeltme: Kanal ekleme sadece TV açıkken yapılabilir
        if not self.tv_acik:
            print("Kanal eklemek için televizyon açık olmalıdır.")
            return

        yeni_kanal = input("Eklemek istediğiniz kanalın ismini giriniz: ")
        if not yeni_kanal: # Boş giriş kontrolü
             print("Kanal ismi boş olamaz.")
             return

        # Büyük/küçük harf duyarsız kontrol ekleyelim
        # Düzeltme: Kontrolü .upper() ile yapalım
        if yeni_kanal.upper() in [name.upper() for name in self.kanallar.values()]:
            print(f"'{yeni_kanal}' kanalı zaten kanal listenizde mevcuttur.")

        else:
            # Bir sonraki boş kanal numarasını bul
            if self.kanallar:
                yeni_key = max(self.kanallar.keys()) + 1
            else:
                yeni_key = 1  # Eğer liste boşsa 1'den başla

            self.kanallar[yeni_key] = yeni_kanal
            print(f"'{yeni_kanal}' kanalı listeye {yeni_key} numarası ile eklenmiştir.")
            print(self.kanallar)
           


    def kanal_sayisi(self):
        # Düzeltme: Kanal sayısı öğrenme sadece TV açıkken yapılabilir
        if not self.tv_acik:
            print("Kanal sayısını öğrenmek için televizyon açık olmalıdır.")
            return

        value_count = len(self.kanallar)
        print(f"Toplam kanal sayısı: {value_count}")  # Daha bilgilendirici çıktı


    def random_kanal(self):
        # Düzeltme: Rastgele kanal sadece TV açıkken yapılabilir
        if not self.tv_acik:
            print("Rastgele kanala geçmek için televizyon açık olmalıdır.")
            return

        if not self.kanallar: # Kanal listesi boşsa kontrol et
            print("Kanal listesi boş. Rastgele kanal seçilemiyor.")
            return # Fonksiyondan çık

        # Rastgele bir anahtar seçip o anahtarın değerini (kanal adını) alalım
        rastgele_kanal_numarasi = random.choice(list(self.kanallar.keys()))
        rastgele_kanal_adi = self.kanallar[rastgele_kanal_numarasi]

        # Mevcut kanalı güncelle (isteğe bağlı)
        self.mevcut_kanal_numarasi = rastgele_kanal_numarasi

        print(f"Rastgele seçilen kanal: {rastgele_kanal_numarasi} - {rastgele_kanal_adi}")
        print(f"Mevcut kanal numarası güncellendi: {self.mevcut_kanal_numarasi}")

    def kanal_degistir(self, kanal_numarasi):
        kanal_numarasi = int(input("Kanal numarasını giriniz: "))
        if kanal_numarasi in self.kanallar:
            self.mevcut_kanal_numarasi = kanal_numarasi
            print(f"Mevcut kanal {self.kanallar[kanal_numarasi]} ({kanal_numarasi}) olarak değiştirildi.")
        else:
            print(f"{kanal_numarasi} numaralı kanal mevcut değil. Lütfen geçerli bir kanal numarası girin.")

        
        
    
            




    def bilgi(self):
        """Televizyonun mevcut durum bilgilerini gösterir."""
        # Düzeltme: Bilgi sadece TV açıkken gösterilebilir
        if not self.tv_acik:
            print("Bilgileri görmek için televizyon açık olmalıdır.")
            return

        print("\n--- Televizyon Bilgileri ---")

        # Mevcut kanal bilgisini ekle (isteğe bağlı)
        mevcut_kanal_adi = self.kanallar.get(self.mevcut_kanal_numarasi, "Bilinmiyor") # Eğer kanal yoksa "Bilinmiyor" yaz
        print(f"Mevcut Kanal: {self.mevcut_kanal_numarasi} - {mevcut_kanal_adi}")

        print(f"Mevcut Ses Seviyesi: {self.volume}")
        print(f"Toplam Kanal Sayısı: {len(self.kanallar)}")

        print("\nKanal Listesi:")
        # Kanalları numarasına göre sıralayarak yazdıralım
        if self.kanallar:
            for numara in sorted(self.kanallar.keys()):
                print(f"  {numara}: {self.kanallar[numara]}")
        else:
            print("  Kanal listesi boş.")

        print("\nKullanılabilir Tuşlar:")
        for tus, aciklama in self.tuslar.items():
            print(f"  '{tus}': {aciklama}")

        print("---------------------------")


# Kumanda nesnesini oluşturma ve çalıştırma kısmı
kumanda_objesi = kumanda()
kumanda_objesi.run()


```
