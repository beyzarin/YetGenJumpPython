# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time
import random
from collections import defaultdict


class kumanda :
    def __init__(self):
        self.kanallar = {1:"TRT ÇOCUK",2:"MİNİKA GO",
                         3:"MİNİKA GO",4:"CARTOON NETWORK",
                         5:"DİSNEY CHANNEL",6:"NİCKELODEON",
                         7:"PLANET ÇOCUK",8:"BABYTV",9:"BBC NEWS",
                         11:"NATİONAL GEOGRAPHİC",12:"NOW TV"}
        self.tuslar = {
                        "p": "TV Aç",
                        "q": "TV KAPAT",
                        "+": "Ses Aç", 
                        "-": "Ses Kıs",
                        "m": "Sessize Al",
                        "u": "Kanal Yukarı",
                        "d": "Kanal Aşağı",
                        "ok": "Onayla / Seç",
                        "esc": "Çık / Geri Dön",
                        "1-9": "Kanal Numarası",
                        "s": "Kaynak Seç (HDMI/AV/USB)",
                        "i": "Bilgi Göster",
                        "g": "Yayın Rehberi",
                        "v": "Kanal sayısı öğrenme",
                        "r": "Rastegele kanala geç",
                        "y": "Kanal Ekle",
                        "b": "Mavi Tuş",
                        ">": "Oynat (▶)",
                        "||": "Durdur (⏸)",
                        ">>": "İleri Sar (⏩)",
                        "<<": "Geri Sar (⏪)"
                           }

        self.volume = 0

    # run fonksiyonu en başa alındı
    def run(self):
        while True:
            self.islem_girisi = input("Yapmak istediğiniz işlemi tuşları kullanarak gerçekleştirin (çıkmak için 'q' tuşuna basın): ")

            # Kullanıcı çıkmak istedi mi?
            if self.islem_girisi == 'q':
                print("Televizyon kapatılıyor.")
                break # Ana while döngüsünü kırar

            # Girilen tuşa göre hangi fonksiyonu çağıracağız?
            action_performed = False # Geçerli bir işlem yapılıp yapılmadığını takip etmek için bayrak

            if self.islem_girisi in ('+', '-'): # Ses açma/kısma
                 self.ses()
                 action_performed = True
            elif self.islem_girisi == 'y': # Kanal ekleme
                 self.kanal_ekleme()
                 action_performed = True
            elif self.islem_girisi == 'v': # Kanal sayısı öğrenme
                 self.kanal_sayısı()
                 action_performed = True
            # Diğer tuşlar için buraya elif blokları ekleyebilirsiniz (örn: 'u', 'd', 'r', 'p' vb.)
            # elif self.islem_girisi == 'p':
            #     print("TV Açılıyor (Fonksiyonu henüz eklemediniz)") # İlgili fonksiyonu çağırın
            #     action_performed = True
            # elif self.islem_girisi == 'r':
            #     print("Rastgele kanala geçiliyor (Fonksiyonu henüz eklemediniz)") # İlgili fonksiyonu çağırın
            #     action_performed = True
            # ...

            # Girilen tuş tanımlı tuşlar arasında mı (genel kontrol)?
            # Bu kısım bilinen ama henüz işlevi tam uygulanmamış tuşlar için mesaj verir.
            if self.islem_girisi in self.tuslar and not action_performed:
                 print(f"'{self.tuslar[self.islem_girisi]}' işlemi seçildi (Henüz tam olarak uygulanmadı).")
                 action_performed = True # Bilinen bir tuş olduğu için geçerli sayalım

            # Eğer yukarıdaki kontrollerden hiçbiri eşleşmediyse (yani 'q' değil ve tanımlı bir tuş değilse)
            if not action_performed:
                 print("Geçersiz işlem! Lütfen belirtilen seçeneklerden birini girin.")
                 print("Mevcut tuşlar:", self.tuslar) # Kullanıcıya tuşları tekrar göster


    def ses(self):
        # Bu iç kontrol kalsın, + ve - arasındaki farkı anlamak için
        if self.islem_girisi == "+" :
            self.volume += 10
            print(f"Ses: {self.volume}") # Daha bilgilendirici çıktı
        elif self.islem_girisi == "-":
            self.volume -= 10
            print(f"Ses: {self.volume}") # Daha bilgilendirici çıktı
        # Not: 'w' ve 'm' tuşları için ses mantığı bu metotta hala eksik, isterseniz ekleyebilirsiniz.


    def kanal_ekleme(self):
        yeni_kanal = input("Eklemek istediğiniz kanalın ismini giriniz: ")

        # Büyük/küçük harf duyarsız kontrol ekleyelim
        if yeni_kanal.upper() in [name.upper() for name in self.kanallar.values()]:
            print(f"'{yeni_kanal}' kanalı zaten kanal listenizde mevcuttur.")

        else:
            # Bir sonraki boş kanal numarasını bul
            if self.kanallar:
                yeni_key = max(self.kanallar.keys()) + 1
            else:
                yeni_key = 1 # Eğer liste boşsa 1'den başla

            self.kanallar[yeni_key] = yeni_kanal
            print(f"'{yeni_kanal}' kanalı listeye {yeni_key} numarası ile eklenmiştir.")


    def kanal_sayısı(self):
        value_count = len(self.kanallar) # values() yerine doğrudan sözlüğün boyutu daha doğru
        print(f"Toplam kanal sayısı: {value_count}") # Daha bilgilendirici çıktı


# Kumanda nesnesini oluşturma ve çalıştırma kısmı
kumanda_objesi = kumanda()
kumanda_objesi.run()
```
